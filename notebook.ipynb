### Merging Flights and Airlines

```python
#in this code, I merged the flights and airlines dataframes to get a more comprehensive view of the flight data along with the airline names. 
# I used a left join to ensure that all records from the flights dataframe are retained. After merging,
# I dropped the redundant IATA_CODE column from the airlines dataframe and renamed columns for clarity.

df_merge_flights_airlines = (pd.merge(df_flights, df_airlines,
                                   left_on='AIRLINE',
                                   right_on='IATA_CODE',
                                   how='left')
                             .drop(columns=['IATA_CODE'])
                             .rename(columns={'AIRLINE_x': 'AIRLINE_CODE',
                                              'AIRLINE_y': 'AIRLINE_NAME'}))
df_merge_flights_airlines.head(5)
```

### Merging Flights and Airports

```python
# In this code, I merged the flights and airports dataframes to include information about the origin airport.
# I used a left join to ensure that all records from the flights dataframe are retained.
# renamed columns for clarity, and dropped the redundant IATA_CODE column from the airports dataframe.
df_origin = (pd.merge(df_merge_flights_airlines, df_airports,
                                   left_on='ORIGIN_AIRPORT',
                                   right_on='IATA_CODE',
                                   how='left',
                                   suffixes=('', '_ORIGIN'))
            .drop(columns=['IATA_CODE'])
            .rename(columns={'ORIGIN_AIRPORT': 'ORIGIN_AIRPORT_CODE',
                             'AIRPORT': 'ORIGIN_AIRPORT_NAME', 
                             'CITY': 'ORIGIN_CITY', 
                             'STATE': 'ORIGIN_STATE', 
                             'COUNTRY': 'ORIGIN_COUNTRY'}
            )
)
df_origin.head(5)
```

```python
# In this code, I merged the flights and airports dataframes to include information about the destination airport.
# I used a left join to ensure that all records from the flights dataframe are retained.
# renamed columns for clarity, and dropped the redundant IATA_CODE column from the airports dataframe.
df_final = (pd.merge(df_origin, df_airports,
                                    left_on='DESTINATION_AIRPORT',
                                    right_on='IATA_CODE',
                                    how='left',
                                    suffixes=('', '_DESTINATION'))
            .rename(columns={'DESTINATION_AIRPORT': 'DESTINATION_AIRPORT_CODE',
                             'AIRPORT': 'DESTINATION_AIRPORT_NAME', 
                             'CITY': 'DESTINATION_CITY', 
                             'STATE': 'DESTINATION_STATE', 
                             'COUNTRY': 'DESTINATION_COUNTRY'}
                    )
            .drop(columns=['IATA_CODE'])
            )
df_final.head(5)
```

```python
def dep_hour_from_sched(x):
    '''
    This function is used to convert HHMM to a 0-23 number
    '''
    
    try:
        v = int(x)
        return(v // 100) % 24
    except Exception:
        return np.nan

df_final['DEP_HOUR'] = df_final['SCHEDULED_DEPARTURE'].apply(dep_hour_from_sched).astype('float32')
df_final.head(5)
```

```python
# Adjusting remaining columns 
# Function that converts the 'HHMM' string to datetime.time
def hhmm_to_timedelta(s: pd.Series) -> pd.Series:
    """
    Converte HHMM (int/str) em Timedelta desde 00:00.
    Trata NaN e 2400 -> 00:00 (sem +1 dia).
    """
    t = pd.to_numeric(s, errors='coerce').astype('Int64')
    is_nan = t.isna()
    t0 = t.where(~t.eq(2400), 0)  # 2400 -> 0000

    hh = (t0 // 100).astype('Int64').fillna(0)
    mm = (t0 % 100).astype('Int64').fillna(0)

    td = pd.to_timedelta(hh, unit='h') + pd.to_timedelta(mm, unit='m')
    td = td.where(~is_nan, pd.NaT)
    return td


def make_scheduled_datetime(df: pd.DataFrame, date_col: str, time_col: str) -> pd.Series:
    """
    Combina FL_DATE (data) + HHMM (agendado) em um datetime.
    Trata NaN e 2400 -> soma +1 dia e usa 00:00.
    """
    t = pd.to_numeric(df[time_col], errors='coerce').astype('Int64')
    is_nan = t.isna()
    is_2400 = t.eq(2400)

    t0 = t.where(~is_2400, 0)
    hh = (t0 // 100).astype('Int64')
    mm = (t0 % 100).astype('Int64')

    base_date = pd.to_datetime(df[date_col], errors='coerce')

    dt = (base_date
          + pd.to_timedelta(hh.fillna(0), unit='h')
          + pd.to_timedelta(mm.fillna(0), unit='m'))

    dt = dt + pd.to_timedelta(is_2400.fillna(False).astype(int), unit='D')
    dt = dt.where(~is_nan & base_date.notna(), pd.NaT)
    return dt
```

```python
chunksize = 1_000_000  

n = len(df_final)
sched_dep_out = []
dep_time_out   = []
sched_arr_out  = []
arr_time_out   = []

for i in trange(0, n, chunksize):
    chunk = df_final.iloc[i:i+chunksize]

    sched_dep_out.append(make_scheduled_datetime(chunk, 'FL_DATE', 'SCHEDULED_DEPARTURE'))
    sched_arr_out.append(make_scheduled_datetime(chunk, 'FL_DATE', 'SCHEDULED_ARRIVAL'))
    
df_final['SCHEDULED_DEPARTURE_DT'] = pd.concat(sched_dep_out, ignore_index=True)
df_final['SCHEDULED_ARRIVAL_DT']   = pd.concat(sched_arr_out,  ignore_index=True)
```

```python
from scipy.stats import pointbiserialr
from sklearn.metrics import roc_auc_score
```

```python
TARGET_COL = "DELAYED"
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
numeric_cols = [c for c in numeric_cols if c != TARGET_COL]

results = []

for col in numeric_cols:
    s = df[[TARGET_COL, col]].dropna()
    if s.empty:
        continue
    
    y = s[TARGET_COL].astype(int).values
    x = s[col].astype(float).values
    
    # --- Point-biserial
    r, p = pointbiserialr(y, x)
    
    # --- AUC (rank-biserial from AUC)
    try:
        auc = roc_auc_score(y, x)
        auc_best = max(auc, 1 - auc)                # best direction
        rank_biserial = 2 * auc_best - 1            # convert AUC to rank-biserial
    except ValueError:
        auc, auc_best, rank_biserial = np.nan, np.nan, np.nan
    
    results.append({
        "feature": col,
        "point_biserial_r": r,
        "p_value": p,
        "auc": auc,
        "auc_best_dir": auc_best,
        "rank_biserial": rank_biserial
    })

results_df = pd.DataFrame(results).sort_values("rank_biserial", ascending=False).reset_index(drop=True)
results_df
```

- Point-biserial correlation (point_biserial_r): Measures the linear relationship between a continuous variable and a binary outcome.
- The rank-biserial correlation is directly related to the AUC of a binary classifier built from a single variable.

```python
from scipy.stats import chi2_contingency


def cramers_v_corrected(confusion_matrix: pd.DataFrame) -> float:
    """Bias-corrected Cramér's V (Bergsma & Wicher, 2013)."""
    chi2, _, _, _ = chi2_contingency(confusion_matrix, correction=False)
    n = confusion_matrix.values.sum()
    r, k = confusion_matrix.shape
    phi2 = chi2 / n
    phi2_corr = max(0, phi2 - ((k - 1)*(r - 1)) / (n - 1))
    r_corr = r - ((r - 1)**2) / (n - 1)
    k_corr = k - ((k - 1)**2) / (n - 1)
    denom = min(r_corr - 1, k_corr - 1)
    return np.sqrt(phi2_corr / denom) if denom > 0 else 0.0

def topk_levels(df, col, k=50, other_label="__OTHER__"):
    """Keep top-K most frequent levels; group the rest into '__OTHER__'."""
    vc = df[col].value_counts(dropna=False)
    keep = set(vc.head(k).index)
    return df[col].apply(lambda v: v if v in keep else other_label)
```

```python
# --- Choose categorical columns
cat_cols = df.select_dtypes(include=["object", "category"]).columns.tolist()
for c in ["DIST_BIN"]:
    if c in df.columns and c not in cat_cols:
        cat_cols.append(c)
cat_cols = [c for c in cat_cols if c != "DELAYED"]

results = []
for col in cat_cols:
    # Optional: cap high-cardinality columns to Top-K levels for stability/speed
    series = df[col]
    if series.nunique(dropna=False) > 200:  # tweak threshold as needed
        series = topk_levels(df, col, k=100)

    # Build contingency table; observed=True drops unused categorical levels
    ct = pd.crosstab(df["DELAYED"], series, dropna=False)

    # Drop empty columns (sum == 0) if any remain
    ct = ct.loc[:, ct.sum(axis=0) > 0]

    # Need at least 2 levels to compute association
    if ct.shape[1] < 2:
        continue

    # Compute Cramér's V (bias-corrected) and chi2 p-value
    v = cramers_v_corrected(ct)
    chi2, p, dof, _ = chi2_contingency(ct, correction=False)

    results.append({
        "feature": col,
        "cramers_v": v,
        "chi2_p_value": p,
        "n_levels": ct.shape[1]
    })

cramers_df = pd.DataFrame(results).sort_values("cramers_v", ascending=False).reset_index(drop=True)
cramers_df
```

- Cramér’s V: Measures the strength of association between a categorical feature and a binary outcome
- Chi² p-value: Tests the null hypothesis that the feature and the target are independent.

```python
from functools import lru_cache
from meteostat import Stations, Hourly

# -------------------- CONFIG --------------------
DF_IN = df  # seu DataFrame
ORIG_LAT, ORIG_LON = "LATITUDE", "LONGITUDE"
DEST_LAT, DEST_LON = "LATITUDE_DESTINATION", "LONGITUDE_DESTINATION"
DT_DEP, DT_ARR = "SCHEDULED_DEPARTURE_DT", "SCHEDULED_ARRIVAL_DT"

# -------------------- PREP ----------------------
df = DF_IN.copy()
df[DT_DEP] = pd.to_datetime(df[DT_DEP])
df[DT_ARR] = pd.to_datetime(df[DT_ARR])

# limpar colunas antigas (idempotente)
for c in ["TZ_ORIG","TZ_DEST","DEP_HOUR_LOCAL","ARR_HOUR_LOCAL","DEP_HOUR_UTC","ARR_HOUR_UTC"]:
    if c in df.columns: df.drop(columns=c, inplace=True)

# ----------------- TIMEZONE (cache) --------------
@lru_cache(maxsize=None)
def tz_from_latlon(lat: float, lon: float) -> str:
    st = Stations().nearby(float(lat), float(lon)).fetch(1)
    return (st.iloc[0]["timezone"] if not st.empty else "UTC") or "UTC"

# map sem merges (_x/_y)
orig_pairs = df[[ORIG_LAT, ORIG_LON]].drop_duplicates().copy()
orig_pairs["TZ_ORIG"] = orig_pairs.apply(lambda r: tz_from_latlon(r[ORIG_LAT], r[ORIG_LON]), axis=1)
orig_map = orig_pairs.set_index([ORIG_LAT, ORIG_LON])["TZ_ORIG"]
df["TZ_ORIG"] = pd.MultiIndex.from_arrays([df[ORIG_LAT], df[ORIG_LON]]).map(orig_map)

dest_pairs = df[[DEST_LAT, DEST_LON]].drop_duplicates().copy()
dest_pairs["TZ_DEST"] = dest_pairs.apply(lambda r: tz_from_latlon(r[DEST_LAT], r[DEST_LON]), axis=1)
dest_map = dest_pairs.set_index([DEST_LAT, DEST_LON])["TZ_DEST"]
df["TZ_DEST"] = pd.MultiIndex.from_arrays([df[DEST_LAT], df[DEST_LON]]).map(dest_map)

# ---------- LOCAL -> UTC (vetorizado, usa 'h') ----------
def localize_block(series_naive, tz_str):
    return (pd.to_datetime(series_naive)
            .dt.tz_localize(tz_str, nonexistent="NaT", ambiguous="NaT")
            .dt.floor("h"))

df["DEP_HOUR_UTC"] = pd.Series(pd.NaT, dtype="datetime64[ns, UTC]", index=df.index)
for tz, g in df.groupby("TZ_ORIG", dropna=False):
    if pd.isna(tz): continue
    loc = localize_block(g[DT_DEP], tz).dt.tz_convert("UTC")
    df.loc[g.index, "DEP_HOUR_UTC"] = loc

df["ARR_HOUR_UTC"] = pd.Series(pd.NaT, dtype="datetime64[ns, UTC]", index=df.index)
for tz, g in df.groupby("TZ_DEST", dropna=False):
    if pd.isna(tz): continue
    loc = localize_block(g[DT_ARR], tz).dt.tz_convert("UTC")
    df.loc[g.index, "ARR_HOUR_UTC"] = loc

# ------------- FETCH (UTC) com estação -------------
def fetch_block_utc(lat, lon, start_utc, end_utc, col_time):
    st = Stations().nearby(float(lat), float(lon)).fetch(1)
    if st.empty or pd.isna(start_utc) or pd.isna(end_utc):
        return pd.DataFrame()
    s0 = pd.Timestamp(start_utc).tz_convert("UTC").tz_localize(None)
    s1 = pd.Timestamp(end_utc).tz_convert("UTC").tz_localize(None)
    if s1 < s0:
        return pd.DataFrame()
    data = Hourly(st, s0, s1, timezone="UTC").fetch().reset_index()
    if data.empty: return data
    data = data.rename(columns={"time": col_time})
    keep = [col_time,"temp","rhum","prcp","snow","wspd","wdir","pres","coco"]
    return data[[c for c in keep if c in data.columns]]

# ---------------- ORIGEM ----------------
wx_chunks = []
for (lat, lon), g in df.groupby([ORIG_LAT, ORIG_LON], dropna=False):
    s = pd.to_datetime(g["DEP_HOUR_UTC"], errors="coerce").dropna()
    if s.empty: continue
    w = fetch_block_utc(lat, lon, s.min(), s.max(), "DEP_HOUR_UTC")
    if not w.empty:
        w.insert(0, ORIG_LAT, lat); w.insert(1, ORIG_LON, lon)
        wx_chunks.append(w)
wx_orig = pd.concat(wx_chunks, ignore_index=True) if wx_chunks else pd.DataFrame()
if not wx_orig.empty:
    df = df.merge(wx_orig, on=[ORIG_LAT, ORIG_LON, "DEP_HOUR_UTC"], how="left")
    df.rename(columns={c:f"{c}_ORIG" for c in ["temp","rhum","prcp","snow","wspd","wdir","pres","coco"] if c in df.columns}, inplace=True)

# ---------------- DESTINO ----------------
wx_chunks = []
for (lat, lon), g in df.groupby([DEST_LAT, DEST_LON], dropna=False):
    s = pd.to_datetime(g["ARR_HOUR_UTC"], errors="coerce").dropna()
    if s.empty: continue
    w = fetch_block_utc(lat, lon, s.min(), s.max(), "ARR_HOUR_UTC")
    if not w.empty:
        w.insert(0, DEST_LAT, lat); w.insert(1, DEST_LON, lon)
        wx_chunks.append(w)
wx_dest = pd.concat(wx_chunks, ignore_index=True) if wx_chunks else pd.DataFrame()
if not wx_dest.empty:
    df = df.merge(wx_dest, on=[DEST_LAT, DEST_LON, "ARR_HOUR_UTC"], how="left", suffixes=("","_DEST"))
    df.rename(columns={c:f"{c}_DEST" for c in ["temp","rhum","prcp","snow","wspd","wdir","pres","coco"] if c in df.columns}, inplace=True)

# ---------------- FLAGS ----------------
for side in ["ORIG","DEST"]:
    df[f"wx_rain_{side}"]  = (df.get(f"prcp_{side}", 0)).fillna(0).gt(0).astype(int)
    df[f"wx_snow_{side}"]  = (df.get(f"snow_{side}", 0)).fillna(0).gt(0).astype(int)
    df[f"wx_windy_{side}"] = (df.get(f"wspd_{side}", 0)).fillna(0).ge(10).astype(int)

print("OK:", df.shape)
df.head(3)
```

```python
df = df.copy()

# 1) Remover duplicatas de timezone (_x/_y)
dup_cols = ["TZ_ORIG_x","TZ_ORIG_y","TZ_DEST_x","TZ_DEST_y"]
df.drop(columns=[c for c in dup_cols if c in df.columns], inplace=True, errors="ignore")

# 2) Recriar flags de clima de forma "NA-aware" (NA se não há dado)
for side in ["ORIG","DEST"]:
    pr  = df.get(f"prcp_{side}")
    sn  = df.get(f"snow_{side}")
    ws  = df.get(f"wspd_{side}")

    df[f"wx_rain_{side}"]  = (pr > 0).astype("Int8")   # fica <NA> onde prcp é NaN
    df[f"wx_snow_{side}"]  = (sn > 0).astype("Int8")
    df[f"wx_windy_{side}"] = (ws >= 10).astype("Int8")

# 3) Indicadores de ausência (úteis pro modelo)
wx_vars = ["temp","rhum","prcp","snow","wspd","wdir","pres","coco"]
for side in ["ORIG","DEST"]:
    for v in wx_vars:
        col = f"{v}_{side}"
        if col in df.columns:
            df[f"{col}_NA"] = df[col].isna().astype("Int8")

# 4) Imputação leve (por aeroporto + mês + hora) com fallback global
def impute_group_median(col, keys):
    med = df.groupby(keys)[col].transform("median")
    df[col] = df[col].fillna(med)
    df[col] = df[col].fillna(df[col].median())  # fallback global

# chaves para origem/destino
df["DEP_MONTH"] = pd.to_datetime(df["SCHEDULED_DEPARTURE_DT"]).dt.month
df["DEP_HOUR"]  = pd.to_datetime(df["SCHEDULED_DEPARTURE_DT"]).dt.hour
df["ARR_MONTH"] = pd.to_datetime(df["SCHEDULED_ARRIVAL_DT"]).dt.month
df["ARR_HOUR"]  = pd.to_datetime(df["SCHEDULED_ARRIVAL_DT"]).dt.hour

for v in ["temp","rhum","prcp","wspd","wdir","pres"]:
    col = f"{v}_ORIG"
    if col in df.columns:
        impute_group_median(col, ["LATITUDE","LONGITUDE","DEP_MONTH","DEP_HOUR"])

for v in ["temp","rhum","prcp","wspd","wdir","pres"]:
    col = f"{v}_DEST"
    if col in df.columns:
        impute_group_median(col, ["LATITUDE_DESTINATION","LONGITUDE_DESTINATION","ARR_MONTH","ARR_HOUR"])

# (opcional) coco/snow têm baixa cobertura: mantenha NA + indicadores
# df["snow_ORIG"].fillna(0, inplace=True); df["snow_DEST"].fillna(0, inplace=True)  # só se quiser

# 5) Conferência rápida
cols_show = ["DEP_HOUR_UTC","ARR_HOUR_UTC",
             "temp_ORIG","prcp_ORIG","wspd_ORIG",
             "temp_DEST","prcp_DEST","wspd_DEST",
             "wx_rain_ORIG","wx_windy_ORIG","wx_rain_DEST","wx_windy_DEST"]
print(df[cols_show].head(3))
```

```python
import numpy as np, pandas as pd
from scipy.stats import spearmanr, chi2_contingency

DF = df  # seu dataframe
TARGET = "DELAYED"
MAX_LEVELS = 300     # evita explosões para categóricas enormes
SAMPLE_PER_CLASS = 100_000  # amostra balanceada p/ velocidade (ajuste/0 = desliga)

# --- 0) amostra balanceada opcional ---
if SAMPLE_PER_CLASS:
    dtmp = (DF.groupby(TARGET, group_keys=False)
              .apply(lambda x: x.sample(min(SAMPLE_PER_CLASS, len(x)), random_state=42))
              .reset_index(drop=True))
else:
    dtmp = DF.copy()

# --- 1) NUMÉRICAS -> Spearman ---
num_cols = dtmp.select_dtypes(include=[np.number]).columns.drop(TARGET, errors="ignore")
rows_num=[]
for col in num_cols:
    x = dtmp[col]
    y = dtmp[TARGET]
    m = x.notna() & y.notna()
    if m.sum() < 100: 
        continue
    r, p = spearmanr(x[m], y[m])
    rows_num.append({"feature": col, "metric": "spearman_r", "score": r, "p_value": p, "n_levels": x.nunique(dropna=True)})

num_table = pd.DataFrame(rows_num)

# --- 2) CATEGÓRICAS -> Cramér’s V + chi² p-value ---
def cramers_v_and_p(ct):
    chi2, p, dof, _ = chi2_contingency(ct, correction=False)
    n = ct.to_numpy().sum()
    phi2 = chi2 / max(n, 1)
    r, k = ct.shape
    # bias-corrected Cramér's V
    phi2corr = max(0, phi2 - (k-1)*(r-1)/(n-1))
    rcorr = r - (r-1)**2/(n-1)
    kcorr = k - (k-1)**2/(n-1)
    v = np.sqrt(phi2corr / max(min(kcorr-1, rcorr-1), 1))
    return float(v), float(p)

cat_cols = dtmp.columns.difference(num_cols.tolist() + [TARGET])
rows_cat=[]
for col in cat_cols:
    x = dtmp[col]
    # transformar qualquer tipo categórico/interval em string rótulo
    if pd.api.types.is_interval_dtype(x) or pd.api.types.is_categorical_dtype(x):
        x = x.astype(str)
    x = x.fillna("__NA__").astype("string")
    nlev = x.nunique()
    if nlev <= 1 or nlev > MAX_LEVELS:
        continue
    ct = pd.crosstab(dtmp[TARGET], x)
    v, p = cramers_v_and_p(ct)
    rows_cat.append({"feature": col, "metric": "cramers_v", "score": v, "p_value": p, "n_levels": int(nlev)})

cat_table = pd.DataFrame(rows_cat)

# --- 3) Ranking combinado ---
rank = (pd.concat([num_table, cat_table], ignore_index=True)
          .assign(abs_score=lambda d: d["score"].abs())
          .sort_values("abs_score", ascending=False)
          .drop(columns="abs_score"))

# visualize top-20
rank.head(20)
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder, FunctionTransformer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
y = df['DELAYED']
X = df.drop(columns=['DELAYED'])

num_cols = X.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = X.select_dtypes(include=['object','category']).columns.tolist()
```

```python
scaler = StandardScaler()
ohe = OneHotEncoder(handle_unknown='infrequent_if_exist',
                    min_frequency=0.03,
                    sparse_output=True,
                    dtype='float32'
                    )
```

```python
pre = ColumnTransformer(
    transformers=[
        ('num', scaler, num_cols),
        ('cat', ohe, cat_cols)
    ],
    remainder='drop',
    sparse_threshold=1,
    verbose_feature_names_out=False
)
```

#### Logistic Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression     
from sklearn.model_selection import StratifiedKFold, GridSearchCV 
from sklearn.metrics import roc_auc_score, average_precision_score, classification_report, confusion_matrix
from sklearn.experimental import enable_halving_search_cv  # noqa: F401
from sklearn.model_selection import HalvingGridSearchCV
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.25, random_state=42)

pipe = Pipeline(steps=[
    ('pre', pre),
    ('clf', LogisticRegression(
        solver='saga',
        penalty='l2',
        C=1.0,
        class_weight='balanced',
        max_iter=500,
        tol=1e-3,
        n_jobs=-1,
    ))
])
```

```python
cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)

param_grid = [
    {"clf__penalty": ["l2"], "clf__C": [0.01, 0.1, 1.0, 10.0]},
    {"clf__penalty": ["l1"], "clf__C": [0.01, 0.1, 1.0, 10.0]}, 
    {"clf__penalty": ["elasticnet"], "clf__C": [0.1, 1.0], "clf__l1_ratio": [0.2, 0.5, 0.8]},
]
```

```python
# HalvingGridSearch was used to gain time due to short period to develop case. 
# If there is more time available, GridSearch is recomended 

grid = HalvingGridSearchCV(
    estimator=pipe,
    param_grid=param_grid,
    scoring='average_precision',
    aggressive_elimination=True,
    refit=True,
    max_resources="auto",  
    cv=cv,
    n_jobs=-1,
    verbose=1
)
```

```python
grid.fit(X_train, y_train)
```

```python
def evaluate_model(name, model, X_test, y_test, plot_curves=True):
    # Trees don't have predict_proba if criterion='poisson'; with gini/log_loss they do.
    y_score = model.predict_proba(X_test)[:, 1]
    ap = average_precision_score(y_test, y_score)
    roc = roc_auc_score(y_test, y_score)
    print(f"\n=== {name} ===")
    print(f"Test AP:  {ap:.4f}")
    print(f"Test ROC: {roc:.4f}")

    # Threshold by F1 on PR
    prec, rec, thr = precision_recall_curve(y_test, y_score)
    f1 = 2 * (prec[:-1] * rec[:-1]) / (prec[:-1] + rec[:-1] + 1e-12)
    ix_f1 = np.nanargmax(f1)
    best_thr = float(thr[ix_f1])
    print(f"Best F1 thr: {best_thr:.4f} | P={prec[ix_f1]:.3f} | R={rec[ix_f1]:.3f} | F1={f1[ix_f1]:.3f}")

    # Report at that threshold
    y_pred = (y_score >= best_thr).astype(int)
    cm = confusion_matrix(y_test, y_pred, labels=[0,1])
    print("Confusion matrix [rows=true, cols=pred]:\n", cm)
    print(classification_report(y_test, y_pred, digits=3))

    if plot_curves:
        # PR
        plt.figure(figsize=(6,4))
        plt.plot(rec, prec, lw=2)
        plt.xlabel("Recall"); plt.ylabel("Precision")
        plt.title(f"{name} - PR (AP={ap:.3f})")
        plt.grid(True, ls="--", alpha=0.5)
        plt.show()

        # ROC
        fpr, tpr, _ = roc_curve(y_test, y_score)
        plt.figure(figsize=(6,4))
        plt.plot(fpr, tpr, lw=2); plt.plot([0,1],[0,1], lw=1)
        plt.xlabel("False Positive Rate"); plt.ylabel("True Positive Rate")
        plt.title(f"{name} - ROC (AUC={roc:.3f})")
        plt.grid(True, ls="--", alpha=0.5)
        plt.show()

    return {"ap": ap, "roc": roc, "best_thr": best_thr}
```

```python
res_interpret = evaluate_model("DecisionTree (depth=3)", tree_interpret, X_test, y_test)
res_perf      = evaluate_model("DecisionTree (depth=8)", tree_perf, X_test, y_test)
```

```python
# ============================================
# LightGBM (GPU) end-to-end: CV + Grid + Final + Eval
# ============================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm

# Sklearn
from sklearn.base import clone
from sklearn.model_selection import StratifiedKFold, ParameterGrid, train_test_split
from sklearn.metrics import (
    average_precision_score, roc_auc_score,
    precision_recall_curve, roc_curve,
    classification_report, confusion_matrix
)

# LightGBM
import lightgbm as lgb
from lightgbm import LGBMClassifier


required = ["X_train", "y_train", "X_test", "y_test"]
missing = [v for v in required if v not in globals()]
if missing:
    raise RuntimeError(f"Missing variables: {missing}. Define them before running.")

# Ensure pandas objects (to use .iloc / column ops consistently)
if not hasattr(X_train, "iloc"): X_train = pd.DataFrame(X_train)
if not hasattr(X_test,  "iloc"): X_test  = pd.DataFrame(X_test)
if not hasattr(y_train, "iloc"): y_train = pd.Series(y_train)
if not hasattr(y_test,  "iloc"): y_test  = pd.Series(y_test)

# Optional: enforce same columns/order between train and test
X_test = X_test[X_train.columns]

# Define your categorical and numeric columns (adapt to your data)
cat_cols = [
    "AIRLINE_NAME","ORIGIN_AIRPORT","DESTINATION_AIRPORT",
    "ORIGIN_STATE","DESTINATION_STATE","ORIGIN_CITY",
    "DESTINATION_CITY","DEP_PERIOD"
]
num_cols = [c for c in X_train.columns if c not in cat_cols]  # auto-fill others as numeric

# Cast categorical dtypes (LGBM will use them natively)
for c in cat_cols:
    if c in X_train.columns:
        X_train[c] = X_train[c].astype("category")
        X_test[c]  = X_test[c].astype("category")

# Optional: align indices
y_train = y_train.reset_index(drop=True)
y_test  = y_test.reset_index(drop=True)
X_train = X_train.reset_index(drop=True)
X_test  = X_test.reset_index(drop=True)

print(f"[OK] X_train: {X_train.shape}, X_test: {X_test.shape}, "
      f"y_train: {y_train.shape}, y_test: {y_test.shape}")
print(f"[INFO] categorical features: {sorted(set(cat_cols) & set(X_train.columns))}")
```

```python

# -------------------------------------------------
# 1) GPU config + base params
# -------------------------------------------------
USE_GPU = True  # set to True for RTX 3070
GPU_PLATFORM = 0
GPU_DEVICE = 0

pos = int(np.sum(y_train == 1))
neg = int(np.sum(y_train == 0))
scale_pos_weight = float(neg) / float(pos)
print(f"[INFO] scale_pos_weight ≈ {scale_pos_weight:.2f} (neg={neg}, pos={pos})")

base_clf_params = dict(
    objective="binary",
    n_estimators=5000,         # large cap; early stopping will cut
    learning_rate=0.01,
    subsample=0.8,             # bagging_fraction
    colsample_bytree=0.8,      # feature_fraction
    reg_lambda=5.0,
    reg_alpha=0.0,
    random_state=42,
    max_depth=([-1, 15, 25]),
    n_jobs=-1,
    scale_pos_weight=scale_pos_weight,
    device_type=("gpu" if USE_GPU else "cpu"),
    gpu_platform_id=GPU_PLATFORM,
    gpu_device_id=GPU_DEVICE,
    max_bin=255,              # try 63 for extra speed (slightly less accurate)
)

# -------------------------------------------------
# 2) Compact grid (FAST) to make it run first
#    You can switch FAST=False later for a larger grid.
# -------------------------------------------------


param_grid = {
        "clf__num_leaves":       [ 127, 255],
        "clf__max_depth":        [-1],
        "clf__learning_rate":    [0.01],
        "clf__min_data_in_leaf": [1000, 5000],
        "clf__reg_lambda":       [0.0, 10.0],
        "clf__max_bin":          [63],     # speed-up on GPU
        # subsample/colsample kept fixed to control runtime
    }


grid = list(ParameterGrid(param_grid))
print(f"[INFO] Grid combos: {len(grid)}")

# -------------------------------------------------
# 3) CV with early stopping (transforms folds!)
#    - Fit "pre" on train fold
#    - Transform X_tr/X_val
#    - Fit LGBM with eval_set on transformed data
# -------------------------------------------------
def cv_score_params(pre, base_clf_params, params, X, y, n_splits=3, early_stopping_rounds=100):
    """Return dict with mean AP/AUC and mean best_iteration across folds."""
    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)
    aps, aucs, best_iters = [], [], []

    # merge params into clf params
    clf_params = base_clf_params.copy()
    # only take "clf__" keys from params
    clf_params.update({k.split("clf__", 1)[1]: v for k, v in params.items() if k.startswith("clf__")})

    for tr_idx, va_idx in skf.split(X, y):
        X_tr, X_val = X.iloc[tr_idx], X.iloc[va_idx]
        y_tr, y_val = y.iloc[tr_idx], y.iloc[va_idx]

        # 1) fit/transform pre on fold
        pre_i = clone(pre)
        # (if you had "pre__*" params in grid, set them here)
        pre_i = pre_i  # placeholder for potential pre params
        X_tr_t = pre_i.fit_transform(X_tr, y_tr)
        X_val_t = pre_i.transform(X_val)

        # 2) fit lgbm with early stopping on transformed data
        clf_i = LGBMClassifier(**clf_params)
        clf_i.fit(
            X_tr_t, y_tr,
            eval_set=[(X_val_t, y_val)],
            eval_metric="auc",
            callbacks=[
                lgb.early_stopping(early_stopping_rounds, verbose=False),
                lgb.log_evaluation(50)
            ]
        )

        # 3) score on validation fold
        y_score = clf_i.predict_proba(X_val_t)[:, 1]
        aps.append(average_precision_score(y_val, y_score))
        aucs.append(roc_auc_score(y_val, y_score))
        best_iters.append(clf_i.best_iteration_ or clf_params.get("n_estimators", 5000))

    return {
        "ap_mean": float(np.mean(aps)),
        "ap_std":  float(np.std(aps)),
        "auc_mean": float(np.mean(aucs)),
        "auc_std":  float(np.std(aucs)),
        "best_iter_mean": int(np.round(np.mean(best_iters))),
    }

# -------------------------------------------------
# 4) Run grid CV
# -------------------------------------------------
results = []
for p in tqdm(grid, desc="Grid search (LightGBM, GPU)"):
    scores = cv_score_params(pre, base_clf_params, p, X_train, y_train,
                             n_splits=3, early_stopping_rounds=100)
    results.append({**p, **scores})

# sort by AP
results = sorted(results, key=lambda d: d["ap_mean"], reverse=True)
best = results[0]
print("\n[BEST - CV]")
for k, v in best.items():
    print(f"{k}: {v}")

# -------------------------------------------------
# 5) Final training with early stopping (10% val split)
# -------------------------------------------------
X_tr, X_val, y_tr, y_val = train_test_split(
    X_train, y_train, test_size=0.10, stratify=y_train, random_state=42
)

# fit final preprocessor
pre_final = clone(pre)
X_tr_t = pre_final.fit_transform(X_tr, y_tr)
X_val_t = pre_final.transform(X_val)

# build final clf params
clf_params_final = base_clf_params.copy()
clf_params_final.update({k.split("clf__",1)[1]: v for k,v in best.items() if k.startswith("clf__")})

clf_final = LGBMClassifier(**clf_params_final)
clf_final.fit(
    X_tr_t, y_tr,
    eval_set=[(X_val_t, y_val)],
    eval_metric="auc",
    callbacks=[lgb.early_stopping(200, verbose=False), lgb.log_evaluation(50)]
)
best_iter_final = clf_final.best_iteration_
print(f"\n[FINAL] best_iteration: {best_iter_final}")

# optional: shrink n_estimators to best_iter for compactness
if best_iter_final is not None:
    clf_final.set_params(n_estimators=best_iter_final)

# assemble final pipeline for inference
from sklearn.pipeline import Pipeline
final_model = Pipeline([("pre", pre_final), ("clf", clf_final)])

# -------------------------------------------------
# 6) Test evaluation (AP/ROC, best F1 thr, report)
# -------------------------------------------------
y_score = final_model.predict_proba(X_test)[:, 1]
ap = average_precision_score(y_test, y_score)
roc = roc_auc_score(y_test, y_score)
print(f"\n[TEST] AP={ap:.4f} | ROC AUC={roc:.4f}")

# Best threshold by F1 on PR
prec, rec, thr = precision_recall_curve(y_test, y_score)
f1 = 2 * (prec[:-1] * rec[:-1]) / (prec[:-1] + rec[:-1] + 1e-12)
ix = np.nanargmax(f1)
best_thr = float(thr[ix])
print(f"[TEST] Best F1 thr={best_thr:.4f} | P={prec[ix]:.3f} | R={rec[ix]:.3f} | F1={f1[ix]:.3f}")

# Confusion matrix & report at F1 thr
y_pred = (y_score >= best_thr).astype(int)
cm = confusion_matrix(y_test, y_pred, labels=[0,1])
print("\n[TEST] Confusion matrix [rows=true, cols=pred]:")
print(cm)
print("\n[TEST] Classification report @F1 thr")
print(classification_report(y_test, y_pred, digits=3))

# PR / ROC curves (optional)
plt.figure(figsize=(6,4))
plt.plot(rec, prec, lw=2)
plt.xlabel("Recall"); plt.ylabel("Precision")
plt.title(f"LightGBM (GPU) - PR (AP={ap:.3f})")
plt.grid(True, ls="--", alpha=0.5)
plt.show()

fpr, tpr, _ = roc_curve(y_test, y_score)
plt.figure(figsize=(6,4))
plt.plot(fpr, tpr, lw=2); plt.plot([0,1],[0,1], lw=1)
plt.xlabel("False Positive Rate"); plt.ylabel("True Positive Rate")
plt.title(f"LightGBM (GPU) - ROC (AUC={roc:.3f})")
plt.grid(True, ls="--", alpha=0.5)
plt.show()

# -------------------------------------------------
# 7) Feature importances (gain) + export
# -------------------------------------------------
feat_names = None
if hasattr(pre_final, "get_feature_names_out"):
    try:
        feat_names = pre_final.get_feature_names_out()
    except Exception:
        pass

imp_split = clf_final.feature_importances_
imp_gain  = clf_final.booster_.feature_importance(importance_type="gain")

if feat_names is None:
    feat_names = np.array([f"feat_{i}" for i in range(len(imp_gain))])

fi_df = pd.DataFrame({"feature": feat_names, "gain": imp_gain, "splits": imp_split}) \
         .sort_values("gain", ascending=False)

print("\n[TOP-30 Feature Importances by GAIN]")
for i, row in fi_df.head(30).iterrows():
    print(f"{row['feature']}: gain={row['gain']:.2f} | splits={row['splits']}")

fi_df.to_csv("lgbm_feature_importances.csv", index=False)
print("\n[INFO] Saved feature importances to lgbm_feature_importances.csv")

# -------------------------------------------------
# 8) Save pipeline
# -------------------------------------------------
import joblib
joblib.dump(final_model, "lightgbm_pipeline_gpu.joblib")
print("[INFO] Saved model to lightgbm_pipeline_gpu.joblib")

# -------------------------------------------------
# 9) Quick GPU confirmation tip:
#    In the training logs you should see: "This is the GPU trainer!!"
#    If not, your LightGBM build isn't GPU-enabled.
# -------------------------------------------------

# -------------------------------------------------
# 1) GPU config + base params
# -------------------------------------------------
USE_GPU = True  # set to True for RTX 3070
GPU_PLATFORM = 0
GPU_DEVICE = 0

pos = int(np.sum(y_train == 1))
neg = int(np.sum(y_train == 0))
scale_pos_weight = float(neg) / float(pos)
print(f"[INFO] scale_pos_weight ≈ {scale_pos_weight:.2f} (neg={neg}, pos={pos})")

base_clf_params = dict(
    objective="binary",
    n_estimators=5000,         # large cap; early stopping will cut
    learning_rate=0.01,
    subsample=0.8,             # bagging_fraction
    colsample_bytree=0.8,      # feature_fraction
    reg_lambda=5.0,
    reg_alpha=0.0,
    random_state=42,
    max_depth=([-1, 15, 25]),
    n_jobs=-1,
    scale_pos_weight=scale_pos_weight,
    device_type=("gpu" if USE_GPU else "cpu"),
    gpu_platform_id=GPU_PLATFORM,
    gpu_device_id=GPU_DEVICE,
    max_bin=255,              # try 63 for extra speed (slightly less accurate)
)

# -------------------------------------------------
# 2) Compact grid (FAST) to make it run first
#    You can switch FAST=False later for a larger grid.
# -------------------------------------------------


param_grid = {
        "clf__num_leaves":       [ 127, 255],
        "clf__max_depth":        [-1],
        "clf__learning_rate":    [0.01],
        "clf__min_data_in_leaf": [1000, 5000],
        "clf__reg_lambda":       [0.0, 10.0],
        "clf__max_bin":          [63],     # speed-up on GPU
        # subsample/colsample kept fixed to control runtime
    }


grid = list(ParameterGrid(param_grid))
print(f"[INFO] Grid combos: {len(grid)}")

# -------------------------------------------------
# 3) CV with early stopping (transforms folds!)
#    - Fit "pre" on train fold
#    - Transform X_tr/X_val
#    - Fit LGBM with eval_set on transformed data
# -------------------------------------------------
def cv_score_params(pre, base_clf_params, params, X, y, n_splits=3, early_stopping_rounds=100):
    """Return dict with mean AP/AUC and mean best_iteration across folds."""
    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)
    aps, aucs, best_iters = [], [], []

    # merge params into clf params
    clf_params = base_clf_params.copy()
    # only take "clf__" keys from params
    clf_params.update({k.split("clf__", 1)[1]: v for k, v in params.items() if k.startswith("clf__")})

    for tr_idx, va_idx in skf.split(X, y):
        X_tr, X_val = X.iloc[tr_idx], X.iloc[va_idx]
        y_tr, y_val = y.iloc[tr_idx], y.iloc[va_idx]

        # 1) fit/transform pre on fold
        pre_i = clone(pre)
        # (if you had "pre__*" params in grid, set them here)
        pre_i = pre_i  # placeholder for potential pre params
        X_tr_t = pre_i.fit_transform(X_tr, y_tr)
        X_val_t = pre_i.transform(X_val)

        # 2) fit lgbm with early stopping on transformed data
        clf_i = LGBMClassifier(**clf_params)
        clf_i.fit(
            X_tr_t, y_tr,
            eval_set=[(X_val_t, y_val)],
            eval_metric="auc",
            callbacks=[
                lgb.early_stopping(early_stopping_rounds, verbose=False),
                lgb.log_evaluation(50)
            ]
        )

        # 3) score on validation fold
        y_score = clf_i.predict_proba(X_val_t)[:, 1]
        aps.append(average_precision_score(y_val, y_score))
        aucs.append(roc_auc_score(y_val, y_score))
        best_iters.append(clf_i.best_iteration_ or clf_params.get("n_estimators", 5000))

    return {
        "ap_mean": float(np.mean(aps)),
        "ap_std":  float(np.std(aps)),
        "auc_mean": float(np.mean(aucs)),
        "auc_std":  float(np.std(aucs)),
        "best_iter_mean": int(np.round(np.mean(best_iters))),
    }

# -------------------------------------------------
# 4) Run grid CV
# -------------------------------------------------
results = []
for p in tqdm(grid, desc="Grid search (LightGBM, GPU)"):
    scores = cv_score_params(pre, base_clf_params, p, X_train, y_train,
                             n_splits=3, early_stopping_rounds=100)
    results.append({**p, **scores})

# sort by AP
results = sorted(results, key=lambda d: d["ap_mean"], reverse=True)
best = results[0]
print("\n[BEST - CV]")
for k, v in best.items():
    print(f"{k}: {v}")

# -------------------------------------------------
# 5) Final training with early stopping (10% val split)
# -------------------------------------------------
X_tr, X_val, y_tr, y_val = train_test_split(
    X_train, y_train, test_size=0.10, stratify=y_train, random_state=42
)

# fit final preprocessor
pre_final = clone(pre)
X_tr_t = pre_final.fit_transform(X_tr, y_tr)
X_val_t = pre_final.transform(X_val)

# build final clf params
clf_params_final = base_clf_params.copy()
clf_params_final.update({k.split("clf__",1)[1]: v for k,v in best.items() if k.startswith("clf__")})

clf_final = LGBMClassifier(**clf_params_final)
clf_final.fit(
    X_tr_t, y_tr,
    eval_set=[(X_val_t, y_val)],
    eval_metric="auc",
    callbacks=[lgb.early_stopping(200, verbose=False), lgb.log_evaluation(50)]
)
best_iter_final = clf_final.best_iteration_
print(f"\n[FINAL] best_iteration: {best_iter_final}")

# optional: shrink n_estimators to best_iter for compactness
if best_iter_final is not None:
    clf_final.set_params(n_estimators=best_iter_final)

# assemble final pipeline for inference
from sklearn.pipeline import Pipeline
final_model = Pipeline([("pre", pre_final), ("clf", clf_final)])

# -------------------------------------------------
# 6) Test evaluation (AP/ROC, best F1 thr, report)
# -------------------------------------------------
y_score = final_model.predict_proba(X_test)[:, 1]
ap = average_precision_score(y_test, y_score)
roc = roc_auc_score(y_test, y_score)
print(f"\n[TEST] AP={ap:.4f} | ROC AUC={roc:.4f}")

# Best threshold by F1 on PR
prec, rec, thr = precision_recall_curve(y_test, y_score)
f1 = 2 * (prec[:-1] * rec[:-1]) / (prec[:-1] + rec[:-1] + 1e-12)
ix = np.nanargmax(f1)
best_thr = float(thr[ix])
print(f"[TEST] Best F1 thr={best_thr:.4f} | P={prec[ix]:.3f} | R={rec[ix]:.3f} | F1={f1[ix]:.3f}")

# Confusion matrix & report at F1 thr
y_pred = (y_score >= best_thr).astype(int)
cm = confusion_matrix(y_test, y_pred, labels=[0,1])
print("\n[TEST] Confusion matrix [rows=true, cols=pred]:")
print(cm)
print("\n[TEST] Classification report @F1 thr")
print(classification_report(y_test, y_pred, digits=3))

# PR / ROC curves (optional)
plt.figure(figsize=(6,4))
plt.plot(rec, prec, lw=2)
plt.xlabel("Recall"); plt.ylabel("Precision")
plt.title(f"LightGBM (GPU) - PR (AP={ap:.3f})")
plt.grid(True, ls="--", alpha=0.5)
plt.show()

fpr, tpr, _ = roc_curve(y_test, y_score)
plt.figure(figsize=(6,4))
plt.plot(fpr, tpr, lw=2); plt.plot([0,1],[0,1], lw=1)
plt.xlabel("False Positive Rate"); plt.ylabel("True Positive Rate")
plt.title(f"LightGBM (GPU) - ROC (AUC={roc:.3f})")
plt.grid(True, ls="--", alpha=0.5)
plt.show()

# -------------------------------------------------
# 7) Feature importances (gain) + export
# -------------------------------------------------
feat_names = None
if hasattr(pre_final, "get_feature_names_out"):
    try:
        feat_names = pre_final.get_feature_names_out()
    except Exception:
        pass

imp_split = clf_final.feature_importances_
imp_gain  = clf_final.booster_.feature_importance(importance_type="gain")

if feat_names is None:
    feat_names = np.array([f"feat_{i}" for i in range(len(imp_gain))])

fi_df = pd.DataFrame({"feature": feat_names, "gain": imp_gain, "splits": imp_split}) \
         .sort_values("gain", ascending=False)

print("\n[TOP-30 Feature Importances by GAIN]")
for i, row in fi_df.head(30).iterrows():
    print(f"{row['feature']}: gain={row['gain']:.2f} | splits={row['splits']}")

fi_df.to_csv("lgbm_feature_importances.csv", index=False)
print("\n[INFO] Saved feature importances to lgbm_feature_importances.csv")

# -------------------------------------------------
# 8) Save pipeline
# -------------------------------------------------
import joblib
joblib.dump(final_model, "lightgbm_pipeline_gpu.joblib")
print("[INFO] Saved model to lightgbm_pipeline_gpu.joblib")

# -------------------------------------------------
# 9) Quick GPU confirmation tip:
#    In the training logs you should see: "This is the GPU trainer!!"
#    If not, your LightGBM build isn't GPU-enabled.
# -------------------------------------------------
```

```python
from sklearn.metrics import precision_recall_curve, confusion_matrix, classification_report

def best_threshold_f2(y_true, y_score, min_precision=None):
    # PR curve
    prec, rec, thr = precision_recall_curve(y_true, y_score)
    prec, rec = prec[:-1], rec[:-1]              # align with thresholds

    # F2 = 5PR / (4P + R)
    f2 = (5.0 * prec * rec) / (4.0 * prec + rec + 1e-12)

    if min_precision is not None:
        valid = np.where(prec >= float(min_precision))[0]
        if len(valid) == 0:
            raise ValueError("No point meets the precision floor")
        idx = valid[np.argmax(f2[valid])]
    else:
        idx = np.argmax(f2)

    return float(thr[idx]), float(prec[idx]), float(rec[idx]), float(f2[idx])

# --- Use: maximize F2 (no precision floor) ---
thr, p, r, f2 = best_threshold_f2(y_test, y_score)
y_pred = (y_score >= thr).astype(int)
print(f"F2-optimal thr={thr:.4f} | P={p:.3f} | R={r:.3f} | F2={f2:.3f}")
print(confusion_matrix(y_test, y_pred, labels=[0,1]))
print(classification_report(y_test, y_pred, digits=3))
```

The model achieves moderate performance, with ROC AUC ≈ 0.68 and Average Precision ≈ 0.12. At the F1-optimal threshold, the model shows limited recall (0.37) and low precision (0.13), which reflects the difficulty of detecting the minority class (delayed flights) in a highly imbalanced dataset.

When shifting the decision threshold to optimize for the **F2-score** (which prioritizes recall over precision), recall increases substantially to **0.62**, while precision drops to **0.096**. This trade-off leads to a lower overall accuracy (0.64), but aligns with the practical objective: **better identification of class 1 (delayed flights), even at the cost of more false positives**.

In contexts where anticipating delays is more important than avoiding false alarms (e.g., operational planning, passenger alerts, resource allocation), the F2-optimal threshold is preferable. The model can serve as a screening tool: flagging a broader set of potentially delayed flights, which can then be further filtered or prioritized using additional rules, business logic, or anomaly-detection signals.
